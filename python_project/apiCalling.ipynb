```python
import requests
from datetime import datetime
r = requests.get("https://www.nepsealpha.com/api/smx9841/investment_calander")
data = r.json()
date = datetime.now()
today_date = date.strftime("%Y-%m-%d") 
data['ipo']
for i in data['ipo']:
    if i['opening_date']==today_date:
        print("Ipo Opening")
        print(i['symbol'])
    if i['closing_date'] ==today_date:
        print("Ipo closing")
        print(i['symbol'])
```
